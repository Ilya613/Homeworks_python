```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime

def habr_searсh_to_dataframe(list_searсh):
  """Функция принимает список запросов для поиска в виде ['запрос1','запрос2','запрос3',...и т.д], 
     выполняе поиск по сайту  habr.com. и возвращает Датафрейм в виде: <дата> - <заголовок> - <ссылка на материал>
     в ответ попадают только первая страницы результатов каждого запроса , содержащие 20 элементов. 
     В конечном датафреме удаляются дубликаты результатов поисковых запросов , если такие были """
  
  url = 'https://habr.com/ru/search/'
  habr_note = pd.DataFrame()
  
  for el in list_searсh:
    params = {'q': el}
    res = requests.get(url,params=params)
    soup = BeautifulSoup(res.text)
    articles = soup.find_all('article',class_='tm-articles-list__item')
    for article in articles:
      try:
        date = article.find('time').get('datetime')
        dt = datetime.datetime.strptime(date, "%Y-%m-%dT%H:%M:%S.000Z")
        title = article.find('h2',class_ = 'tm-article-snippet__title tm-article-snippet__title_h2').text
        link = 'https://habr.com' + article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').find('a').get('href')
      except Exception:
        continue
      row = {'date': dt.date(), 'title': title, 'link': link}
      habr_note = pd.concat([habr_note, pd.DataFrame([row])]) 
  habr_note.drop_duplicates('link', inplace=True)
  habr_note.reset_index(inplace=True)
  habr_note.pop('index')
  return habr_note


habr_searсh_to_dataframe(['python','python в мире','анализ данных'])
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime
import time

def habr_searсh_to_dataframe_v2(list_searсh,number_of_pages):
  """Функция принимает список запросов для поиска в виде ['запрос1','запрос2','запрос3',...и т.д],
     и ЧИСЛО  являющееся количеством страниц поисковой выдачи, поиск выполняется по сайту  habr.com. 
     Возвращает Датафрейм в виде: <дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков>
     В конечном датафреме удаляются дубликаты результатов поисковых запросов , если такие были """

  habr_note = pd.DataFrame()  
  for el in list_searсh:
    url = 'https://habr.com/ru/search/'
    page = 1
    while page < number_of_pages+1:
      params = {'q': el}
      res = requests.get(url,params=params)
      time.sleep(0.33)
      soup = BeautifulSoup(res.text)
      articles = soup.find_all('article',class_='tm-articles-list__item')

      for article in articles:
        try:
          date = article.find('time').get('datetime')
          dt = datetime.datetime.strptime(date, "%Y-%m-%dT%H:%M:%S.000Z")
          title = article.find('h2',class_ = 'tm-article-snippet__title tm-article-snippet__title_h2').text
          link = 'https://habr.com' + article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').find('a').get('href')
        except Exception:
          continue
        row = {'date': dt.date(), 'title': title, 'link': link}
        habr_note = pd.concat([habr_note, pd.DataFrame([row])])
      page = page+1
      url_page = 'page'+ str(page)
      url = f'https://habr.com/ru/search/{url_page}/'

  habr_note.drop_duplicates('link', inplace=True)
  habr_note.reset_index(inplace=True)
  habr_note.pop('index')

  i = 0
  for el in habr_note['link']:
    try:
      req = requests.get(el)
      time.sleep(0.33)
      soup2 = BeautifulSoup(req.text)
      full_text = soup2.find('div', class_ = 'article-formatted-body article-formatted-body article-formatted-body_version-1').text
    except AttributeError:
      try:
        req = requests.get(el)
        time.sleep(0.33)
        soup2 = BeautifulSoup(req.text)
        full_text = soup2.find('div',class_ = 'article-formatted-body article-formatted-body article-formatted-body_version-2').text
      except AttributeError:
        full_text = 'Нет текста'
    try:
      req_like = requests.get(el)
      time.sleep(0.33)
      soup_like = BeautifulSoup(req.text)
      likes = soup_like.find('div', class_ = 'tm-votes-meter tm-article-rating__votes-switcher').text
    except Exception:
      i += 1
      continue  
    habr_note.loc[i, 'text'] = full_text
    habr_note.loc[i, 'like_info'] = likes
    i += 1

  return habr_note

habr_searсh_to_dataframe_v2(['python','анализ данных','парсинг'], 3) 
```
