```python
import numpy as np
```

```python

trials = 1000000  # Количество попыток
result_777 = 0
result_999 = 0
result_555 = 0
result_333 = 0
result_111 = 0
result_77 = 0
result_7 = 0
result_00 = 0
result_0 = 0

for i in range(trials):
    random_score = np.random.randint(0, 1000)
    if random_score == 999:
        result_999 += 1
    elif random_score == 777:
        result_777 += 1         
    elif random_score == 555:
        result_555 += 1
    elif random_score == 333:
        result_333 += 1  
    elif random_score == 111:
        result_111 += 1  
    elif random_score % 100 == 0 and random_score !=0 :
        result_00 += 1  
    elif random_score % 10 == 0:
        result_0 += 1 
    elif random_score % 10 == 7 and random_score % 100 != 77:
        result_7 += 1
    elif random_score % 100 == 77:
        result_77 += 1

total_win = ((result_777*200) + (result_999*100) + (result_555*50) + (result_333*15) + 
             (result_111*10) + (result_77*5) + (result_7*3) + (result_00*2) + (result_0*1)) 

win_combinations = (result_777 + result_999 + result_555 + result_333 + 
                    result_111 + result_77 + result_7 + result_00 + result_0)
average_payoff = round((total_win / win_combinations),2)

print(f'777 выпало - {result_777} раз')
print(f'999 выпало - {result_999} раз')
print(f'555 выпало - {result_555} раз')
print(f'333 выпало - {result_333} раз')
print(f'111 выпало - {result_111} раз')
print(f'*77 выпало - {result_77} раз')
print(f'**7 выпало - {result_7} раз')
print(f'*00 выпало - {result_00} раз')
print(f'**0 выпало - {result_0} раз')

print(f'Кол-во попыток {trials}, Кол-во выигрышных комбинаций {win_combinations}, Вероятность выигрыша {win_combinations/trials}')
print(f'Всего выплат по всем выигрышам {total_win} руб')
print()
print(f'За одну игру игрок либо теряет 1 рубль в случае проигрыша ,\nлибо выигрывает в среднем {average_payoff-1} руб')

# Игра Игроку не выгодна .
```
