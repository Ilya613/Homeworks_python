```python
import pandas as pd

ratings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Pandas/Pandas1/ratings.csv')
```

```python
def rating_class(rating):
  if rating <= 2:
    return 'низкий рейтинг'
  elif 2 < rating <= 4:
    return 'средний рейтинг'
  elif 4 < rating <= 5:
    return 'высокий рейтинг'


ratings['class'] = ratings['rating'].apply(rating_class)
ratings.head()
```

```python
 keywords = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Pandas/Pandas_2/ml-latest-small/keywords.csv')

 geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def add_region(row_keyword):
  for key, val in geo_data.items():
    for el in val:
      if el in row_keyword.split( ):
        return key
  else:
    return 'undefined'

keywords['region'] = keywords['keyword'].apply(add_region)

keywords.groupby('region').count().reset_index()
```

```python
years = [i for i in range(1950, 2011)]

films = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Pandas/Pandas1/movies.csv')
```

```python
def production_year(text):
  for year in years:
    if str(year) in text:
      return year
  else:
    return 1900    

films['year'] = films['title'].apply(production_year)
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Pandas/Pandas1/ratings.csv')
```

```python
def average_rating(data):
    return data.rating.mean()

dict_rating = ratings.groupby('movieId').apply(average_rating).to_dict()

def add_average_rating(data):
  return dict_rating.get(data)


films['average_rating'] = films['movieId'].apply(add_average_rating)
```

```python
result = films.groupby('year').agg(['mean'])['average_rating']
result.sort_values('mean', ascending=False)
```
