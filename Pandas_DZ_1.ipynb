```python
import pandas as pd

ratings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Pandas/Pandas1/ratings.csv')

rating_5 = ratings[ ratings.rating == 5 ]
result_id = rating_5['movieId'].value_counts().head(1)
result_id_film = str(result_id).split(' ')[0]

f = open('/content/drive/MyDrive/Colab Notebooks/Pandas/Pandas1/movies.csv', 'r', encoding='utf-8')
film_list = f.readlines()
for i in film_list:
  film = i.split(',', 1)
  if result_id_film == film[0]:
    print(f'Фильму с ID {result_id_film} выставлено больше всего оценок 5.0. \n Название фильма: {film[1]}')


```
