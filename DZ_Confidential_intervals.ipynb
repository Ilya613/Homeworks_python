Задание № 2
Даны две выборки роста мужчин и женщин.

Докажите, используя t-Тест Стьдента, что различие между выборками незначительно, если уровень значимости равен  α  = 0.001.
Покажите различия визуально

```python
import numpy as np  
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)

mens  = stats.norm.rvs(loc=171, scale=100, size=150000)  # Выборка мужчин со средним ростом 171
womens = stats.norm.rvs(loc=170, scale=100, size=150000)  # Выборка женщин со средним ростом 170
```

```python
stats.ttest_ind(mens, womens, equal_var = False) #Критерий Стьюдента (t-test)
```

Вывод : так как p_value меньше уровеня значимости( α = 0.001), то мы отвергаем нулевую гипотезу и значит что различия между выборками есть

```python
plt.hist(mens, bins=150);
plt.hist(womens, bins=150); 
```

Задание № 5

Используя функцию mean_confidence_interval(data, confidence), постройте доверительный интервал с уровнем доверия  β=90%  для выборки:

```python
data = [4,5,8,9,6,7,5,6,7,8,5,6,7,0,9,8,4,6,
        7,9,8,6,5,7,8,9,6,7,5,8,6,7,9,5,10]

def mean_confidence_interval(data, confidence=0.95):
    n = len(data)
    m, se = np.mean(data), stats.sem(data)
    h = se * stats.t.ppf((1 + confidence)/2, n)
    return m-h, m, m+h 

mean_confidence_interval(data, 0.90)    
```
